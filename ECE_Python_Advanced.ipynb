2 b) Printing all the primes less than a given number

```python
# Python3 program to print 
# all primes less than N
  
# Function to check whether 
# a number is prime or not .
def isPrime(n):
      
    # Corner case
    if n <= 1 :
        return False
  
    # check from 2 to n-1
    for i in range(2, n):
        if n % i == 0:
            return False
  
    return True
  
# Function to print primes
def printPrime(n):
    for i in range(2, n):
        if isPrime(i):
            print(i, end = " ")
  
# Driver code
if __name__ == "__main__" :
    n = int(input('Enter the number to print the prime numbers:'))
    # function calling
    printPrime(n)
```

2 c) Finding all the factors of a number and show whether it is a perfect number, i.e., the sum of all its factors (excluding the number itself) is equal to the number itself

```python
# Function to determine if given number is a perfect number
# Input: Number (Range: [1, sys.maxsize])
# Output: A string detailing if given number is perfect or not
def Perfect_Number(num):   
  count=num-1             # count will be our iterator and facilitate our exit from the loop when needed (i.e. when its value = 0)
  sum=0 
  print('The factors of',num,'are')                  # our tracking variable sum
  while count!=0:         
    if num%count == 0:    # when num is divisible by count with 0 as remainder, it means count is a factor of num
      sum += count        # adding factor to the sum
      print(count)
    count-=1                 # decrementing count to explore the next number
  if sum == num:          # Comparing sum with num, to get the verdict of perfect/non-perfect number
    print(num,"is a perfect number")
  else:
    print(num,"is not a perfect number")

#Invoking the function
#PerfectNumbers
n = int(input('Enter the number to print the prime numbers:'))
Perfect_Number(n)         # Calling the function and giving n as input to check whether it is a perfect number or not
```

b) Define a boolean function is palindrome(<input>)

```python
# Base Case: '' => True
# Recursive Case: if first and last characters don't match => False
# if they do match, is the middle a palindrome?

def is_palindrome(s):
    if s == "":
        return True
    else:
        if s[0] == s[-1]:
            return is_palindrome(s[1:-1])
        else:
            return False
string = input('Enter the string:')
print(is_palindrome(string))
```

c) Write a function collatz(x) which does the following: if x is odd, x = 3x + 1; if x is even,
then x = x/2. Return the number of steps it takes for x = 1


```python
# Python 3 program to print
# Collatz sequence
 
def collatz(n):
     
    # We simply follow steps
    # while we do not reach 1
    no_of_steps=0
    while n != 1:
      
        print(n, end = ' ')
 
        # If n is odd
        if n & 1:
            n = 3 * n + 1
            no_of_steps=no_of_steps+1
 
        # If even
        else:
            n = n // 2
            no_of_steps=no_of_steps+1
 
    # Print 1 at the end
    print(n)
    print("The number of steps:",no_of_steps+1)
 
# Driver code
n = int(input('Enter the positive integer number either even or odd:'))
collatz(n)
```
